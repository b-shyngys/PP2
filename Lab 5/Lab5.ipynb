1. Write a Python program that matches a string that has an 'a' followed by zero or more 'b''s.

```python
from re import *
with open("row.txt", "r", encoding="utf-8") as file:
    info = file.read().splitlines() 

patt = "ab+"
for item in info:
    if search(patt,item):
        print(item)
```

2.Write a Python program that matches a string that has an 'a' followed by two to three 'b'.

```python
patt = "ab{2,3}"
for item in info:
    if search(patt,item):
        print(item)
```

3.Write a Python program to find sequences of lowercase letters joined with a underscore.

```python
patt = r"[a-z]+_[a-z]+"
for item in info:
    if search(patt,item):
        print(item)
```

5.Write a Python program that matches a string that has an 'a' followed by anything, ending in 'b'.

```python
patt = r"a.*b"
for item in info:
    if match(patt,item):
        print(item)
```

7.Write a python program to convert snake case string to camel case string.

```python
info7 = info.copy()
for item in range(len(info7)):
    snake = info7[item].split()
    for i in range(len(snake)):
        if search("_",snake[i]):  
            snakelist = snake[i].split("_") 
            snakelist = [word.capitalize() for word in snakelist]  
            info7[item] = "".join(snakelist)
    
  
print(info7)
            
```

8.Write a Python program to split a string at uppercase letters.

```python
patt = r"[A-Z]"
info8=info
for item in info8:
    print(split(patt,item))
```

9.Write a Python program to insert spaces between words starting with capital letters.

```python
patt = r"([A-Z])"
info9=info
for item in info9:
    if search(patt,item):
        print(sub(patt,r" \1",item))
```

10.Write a Python program to convert a given camel case string to snake case.

```python
info10 = info.copy()
patt = r"([A-Z])"
for item in range(len(info10)):
    camel = info10[item].split()  
    for i in range(len(camel)): 
        if search(patt,camel[i]): 
            camel[i] = sub(patt,r"_\1",camel[i]).lower()
            camel[i] = camel[i][1:]
    info10[item] = " ".join(camel) 
print(info10)
```
